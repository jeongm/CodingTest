```python
def nsum(n):
    if n == 1:
        return 1
    else :
        return n + nsum(n-1)
```

```python
print(nsum(10))
```

### n개의 원소 중 m개를 고르는 모든 조합을 찾는 알고리즘
입력 : 7 4  
출력: (0,1,2,3), (0,1,2,4), (0,1,2,5)...

```python
n = int(input())
m = int(input())
```

```python
def pick(n,m,a):
    
    if m == 0: # 기저 사례 더 고를 원소가 없으므로 출력
        print(a)
    if not a: # 고른 원소가 없을 때
        point = 0 # 원소 시작
    else :
        point = a[-1]+1 # 가장 마지막에 들어있는 원소 다음 번호부터 시작
        
    for i in range(point, n):
        a.append(i)
        
    
```

```python
a = []
pick(n,m,a)
```

```python
def pick(n, m, picked):

    if m == 0:
        print(picked)

    if not picked:
        last = 0
    else :
        last = picked[len(picked)-1]+1

    for i in range(last, n):
        picked.append(i)
        print("들어감",n,m,i,picked)
        pick(n, m - 1, picked)
        print("나감",n,m,i,picked)
        del picked[len(picked) - 1]
        print("지움",n,m,i,picked)
if __name__ == '__main__':
    a = []
    n = int(input())
    m = int(input())
    pick(n, m, a)

```
